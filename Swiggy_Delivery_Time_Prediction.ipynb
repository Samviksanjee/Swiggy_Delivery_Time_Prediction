__Outliers__

````python
df['age'].plot(
    kind='box',
    vert=True,               # vertical box (default = True)
    patch_artist=True,       # fill with color
    boxprops=dict(facecolor='lightblue', color='black'),  # box color & border
    medianprops=dict(color='red', linewidth=2),            # median line style
    figsize=(6,6),
    title='Boxplot of Age'
)

plt.ylabel('Age')
plt.show()
````



### 🔍 **2️⃣ What Each Part Does**

| Code Component               | Meaning                                                                 |
| ---------------------------- | ----------------------------------------------------------------------- |
| `df['age'].plot(kind='box')` | Creates a **boxplot** of the `age` column.                              |
| `vert=True`                  | Makes the box **vertical** (default).                                   |
| `patch_artist=True`          | Enables coloring inside the box (otherwise, it’s just outlined).        |
| `boxprops`                   | Defines the box’s **style** — here, a light blue fill with black edges. |
| `medianprops`                | Defines how the **median line** inside the box looks — red and thick.   |
| `figsize=(6,6)`              | Sets the overall figure size (6×6 inches).                              |
| `title='Boxplot of Age'`     | Adds a title at the top.                                                |
| `plt.ylabel('Age')`          | Labels the y-axis.                                                      |
| `plt.show()`                 | Displays the final plot.                                                |

---

### 📊 **3️⃣ How to Read a Boxplot**

A boxplot helps you see the **spread, central value, and outliers** in a numerical column like `age`.

#### Components:

* **Box (light blue)** → covers the **middle 50% of data** (from Q1 to Q3).
* **Red line** inside the box → the **median age** (middle value).
* **Whiskers** → extend to the minimum and maximum *non-outlier* values.
* **Dots (if shown)** → represent **outliers**, i.e., unusually high or low ages.

---

### 📈 **4️⃣ Interpreting What You See**

Let’s say the plot looks like this (hypothetically):

* The **box is centered around 25–35 years** → most riders/customers are in that age range.
* The **median (red line)** is near 30 → that’s the typical or middle age.
* The **whiskers extend from 18 to 50** → that’s roughly the total age spread.
* A few **dots above 60** → older customers are rare (outliers).

---

### 💡 **5️⃣ Why Use Boxplots**

Boxplots are great for:

* Quickly detecting **outliers**.
* Comparing **spread and symmetry** of data.
* Understanding the **central tendency** (median).
* Comparing **multiple numerical columns** side-by-side.

---

Would you like me to show how to **create multiple boxplots together** (for example: `age`, `ratings`, `distance`, and `time_taken`) in one figure for comparison?


````python
df['ratings'].plot(
    kind='box',
    vert=True,               # vertical box (default)
    patch_artist=True,       # fill the box with color
    boxprops=dict(facecolor='lightyellow', color='black'),
    medianprops=dict(color='red', linewidth=2),
    whiskerprops=dict(color='black'),
    capprops=dict(color='black'),
    flierprops=dict(marker='o', color='red', markersize=5),
    figsize=(6,6),
    title='Boxplot of Ratings'
)

plt.ylabel('Ratings')
plt.show()
````

## 🔍 2️⃣ What Each Part Does

| Parameter               | Purpose                                                            |
| ----------------------- | ------------------------------------------------------------------ |
| `kind='box'`            | Creates a boxplot for the `ratings` column.                        |
| `vert=True`             | Keeps the box **vertical**.                                        |
| `patch_artist=True`     | Allows **filling color** inside the box.                           |
| `boxprops`              | Defines **color & border** for the box.                            |
| `medianprops`           | Styles the **median line** — here it’s red and thicker.            |
| `whiskerprops`          | Styles the **lines extending** from the box (showing data spread). |
| `capprops`              | Styles the small **caps** at the ends of whiskers.                 |
| `flierprops`            | Styles the **outliers** (data points far from the rest).           |
| `figsize`               | Sets the chart’s width and height.                                 |

---

## 🎯 What is R² (Coefficient of Determination)?

**R²** tells us **how well our regression model explains the variation** in the target variable (for you, it’s `time_taken`).

In simple words:

> R² measures how much of the **change in the output** can be explained by the **input features**.

---

## 🧮 Formula

[
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
]

Where:

* **SS_res (Residual Sum of Squares):**
  [
  SS_{res} = \sum_{i=1}^{n}(y_i - \hat{y_i})^2
  ]
  → It measures how far the predicted values ((\hat{y_i})) are from the actual values ((y_i)).

* **SS_tot (Total Sum of Squares):**
  [
  SS_{tot} = \sum_{i=1}^{n}(y_i - \bar{y})^2
  ]
  → It measures how far the actual values are from the mean of actual values ((\bar{y})).

---

## 🧠 Intuitive Understanding

| Term       | Meaning                                                    |
| ---------- | ---------------------------------------------------------- |
| **SS_tot** | Total variation in the data (what we’re trying to explain) |
| **SS_res** | Variation the model **failed** to explain (errors)         |
| **R²**     | Fraction of variation the model **successfully explained** |

So if:
[
SS_{res} = 0
]
That means predictions are **perfect** — the model explained **100% of the variation**, hence **R² = 1**.

If:
[
SS_{res} = SS_{tot}
]
That means the model did **no better than just predicting the mean**, hence **R² = 0**.

If:
[
SS_{res} > SS_{tot}
]
That means the model is **worse than predicting the mean** — usually happens when the model fits noise → **R² is negative**.

---

## 🔢 Example

Let’s say:

| Metric | Value |
| ------ | ----- |
| SS_tot | 100   |
| SS_res | 40    |

Then:
image.png

✅ Interpretation:
→ The model explains **60% of the variation** in delivery time.
→ The remaining **40%** is unexplained noise or randomness.

| **1.0**       | Perfect model (no errors)                    |

| **0 – 0.4**   | Weak fit                                     |

### 🌲 **Random Forest Regressor — Concept Recap**

A **Random Forest** is an ensemble (a collection) of **many Decision Trees**, where:

* Each tree is trained on a *random subset of the data* (called bootstrapping).
* Each tree considers *a random subset of features* at every split.
* The final prediction is the **average** of all the trees’ predictions.

This reduces **overfitting** and improves **generalization** compared to a single Decision Tree.

## Compare the performance of all models

## Step 1: Store all model results

````python
from sklearn.metrics import r2_score, mean_squared_error
import pandas as pd
import matplotlib.pyplot as plt

# Create a results dictionary
results = {
    "Model": ["Linear Regression", "KNN Regressor", "Support Vector Regressor", "Decision Tree", "Random Forest"],
    "MSE": [],
    "R2 Score": []
}

# Define models
models = [
    ("Linear Regression", Model_lr),
    ("KNN Regressor", Model_knn),
    ("Support Vector Regressor", Model_svr),
    ("Decision Tree", Model_dt),
    ("Random Forest", Model_rf)
]

# Evaluate each model
for name, model in models:
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results["MSE"].append(mse)
    results["R2 Score"].append(r2)

# Convert to DataFrame
results_df = pd.DataFrame(results)
results_df
````

Step 2: Plot Model Comparison

````python
plt.figure(figsize=(10,5))

# R2 Score comparison
plt.subplot(1,2,1)
plt.bar(results_df["Model"], results_df["R2 Score"], color='lightgreen')
plt.title("R² Score Comparison")
plt.ylabel("R² Score")
plt.xticks(rotation=45)

# MSE comparison
plt.subplot(1,2,2)
plt.bar(results_df["Model"], results_df["MSE"], color='lightcoral')
plt.title("Mean Squared Error Comparison")
plt.ylabel("MSE")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
````

__Step 3: Interpretation__

- Higher R² Score → Better model performance.

- Lower MSE → Better accuracy (smaller prediction error).

__Usually:__

- Random Forest performs best (high R², low MSE)

- Linear Regression performs worst if relationships are nonlinear

- SVM and KNN vary depending on data scale and distribution
